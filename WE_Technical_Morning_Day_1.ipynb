```python
def largest(int1 : int, int2 : int, int3 : int)->int:

  if(int1 >= int2 and int1>= int3):
    return int1
  elif(int2 >= int1 and int2 >= int3):
    return int2
  elif (int3 >=int1 and int3 >=int2):
    return int3
```

```python
largest(-1,-2,-1)
```

```python
def larger (a: int, b: int)-> int:
    return a if a> b else b

def largest (a : int, b: int,c: int) -> int:
    return larger(a, larger(b,c))
```

```python
def largest1(a1,b1,c1)-> int:
    largest = a1
    if largest < b1:
        largest = b1
    if largest < c1:
        largest = c1
    return largest
```

```python
def next_palindrome(num1):

    flag = 1

    check = num1+1

    while flag:

        if str(check) == str(check)[::-1]:
            flag = 0
            return check

        check+=1
```

```python
next_palindrome(123)
```

```python
def next_higher_palindrome(n: int) -> int:
    flag = 0
    while flag == 0:
        n += 1
        if str(n) == str(n)[::-1]:
            flag = 1
    return n
```

```python
next_higher_palindrome(123)
```

```python
def unique_sum(int_list):

    freq = {}
    total = 0
    for item in int_list:
        if item not in freq:
            freq[item] = 1
        else:
            freq[item] += 1

    for pair in freq.items():
        if pair[1] ==1:
            total += pair[0]


    return(total)
```

```python
unique_sum([1,2,2,4])
```
