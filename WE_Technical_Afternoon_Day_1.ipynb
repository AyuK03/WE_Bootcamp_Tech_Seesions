```python
def last_digit(n : int) ->int:
    return n % 10

def is_ascending(n : int) -> bool:

    while n > 9:         # IMP LINE

        if last_digit(n) <= last_digit( n // 10 ):

            return False

        n //= 10

    return True
```

#COLLATZ Sequence

```python
def collatz(num : int) -> list[int]:

    collatz_seq = [num]

    while num != 4:

        if num % 2 ==0:
            num = num // 2
        else:
            num = ( 3 * num ) + 1

        collatz_seq.append(num)

    return collatz_seq + [ 2, 1 ]

```

```python
def next_collatz(n: int) -> int:
    if n % 2 == 0:
        return n // 2
    else:
        return 3 * n + 1

def collatz_sequence(n: int) -> list[ int ]:

    seq = []
    while n != 4:
        seq.append(n)
        n = next_collatz(n)
    return seq + [4,2,1]
```

```python
def min_max_diff(n : int)-> int:

    seq = str(n)
    smallest = "".join(sorted(seq))          #smallest must be subscriptable -> so it must be a string
    largest = int("".join(smallest[::-1]))
    return (largest-(int(smallest)))

def kaprekar_sequence(num: int)-> list[int]:

    kaprekar_list = []

    num = min_max_diff(num)

    while num != min_max_diff(num):
        kaprekar_list.append(num)
        num = min_max_diff(num)

    kaprekar_list.append(num)
    return kaprekar_list

```

```python
kaprekar_sequence(5778)
```
