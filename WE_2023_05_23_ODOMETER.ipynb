# Odometer

```python
def size(n: int) -> int:
    return len(str(n))

def get_limits(n: int) -> tuple[int, int]:
    DIGITS = "123456789"
    k = size(n)
    return int(DIGITS[:k]), int(DIGITS[-k:])
```

```python
# THIS IS A SMART CODE

def is_ascending(n: int) -> bool:
    if n < 10:                      # that is, it has reached the last digit
        return True
    if (n % 10) <= (n//10) % 10:
        return False
    return is_ascending(n//10)
```

```python
get_limits(67890)
```

```python
def next_reading(n: int) -> int:
    START, LIMIT = get_limits(n)
    if n == LIMIT:
        return START
    n += 1
    while not is_ascending(n):
        n += 1
    return n
```

```python
def prev_reading(n: int) -> int:
    START, LIMIT = get_limits(n)
    if n == START:
        return LIMIT
    n -= 1
    while not is_ascending(n):
        n -= 1
    return n
```

```python
def distance(a_reading: int, b_reading: int) -> int:
    if size(a_reading) != size(b_reading):  # if the odometer's no. of digits being displayed is different -> not a valid odometer
        return -1
    steps = 0
    while a_reading != b_reading:
        steps += 1
        a_reading = next_reading(a_reading)
    return steps
```

```python
def next_kth_reading(n: int, k: int) -> int:
    for _ in range(k):
        n = next_reading(n)
    return n
```

```python
def prev_kth_reading(n: int, k: int) -> int:
    for _ in range(k):
        n = prev_reading(n)
    return n
```

## Alternately,

```python
def after(n: int, k = 1) -> int:
    START, LIMIT = get_limits(n)
    for _ in range(k):
        if n == LIMIT:
            n = START
        else:
            n += 1
            while not is_ascending(n):
                n += 1
    return n
```

```python
def before(n: int, k = 1) -> int:
    START, LIMIT = get_limits(n)
    for _ in range(k):
        if n == START:
            n = LIMIT
        else:
            n -= 1
            while not is_ascending(n):
                n -= 1
    return n
```
